```python
import numpy as np
import pandas as pd
```

Инициализируем исходные данные по задаче (вариант 23).

```python
m = 8
p = 12
f = pd.Series([12, 10, 8, 6, 4, 2, 0, 0, 0], name='f_from_t')
f
```

```python
frame = pd.concat({(m, 'B'): f.copy(), (m, 'x'): f.apply(lambda x: 0 if x > 0 else (0,5 if x == 0 else 1))}, axis=1)
frame.index.name = 't'
frame.columns.names = ['i', 'step_vals']
frame
```

```python
def get_B_val(row: pd.Series):
    if row.name == m:
        next_B = 0
    else:
        next_B = frame.iloc[row.name + 1, -2]

    return max(f[row.name] + next_B, frame.iloc[1, -2])

def get_x_val(row: pd.Series):
    if row.name == m:
        next_B = 0
    else:
        next_B = frame.iloc[row.name + 1, -2]

    if f[row.name] + next_B > frame.iloc[1, -2]:
        return 0
    elif f[row.name] + next_B < frame.iloc[1, -2]:
        return 1
    else:
        return 0,5

for i in np.arange(m-1, 0, -1):
    step_B = []
    step_x = []
    for t in np.arange(m + 1):
        step_B.append(get_B_val(frame.loc[t]))
        step_x.append(get_x_val(frame.loc[t]))
    frame = frame.merge(pd.Series(step_B, name=(i, 'B')), left_index=True, right_index=True)
    frame = frame.merge(pd.Series(step_x, name=(i, 'x')), left_index=True, right_index=True)
frame
```
